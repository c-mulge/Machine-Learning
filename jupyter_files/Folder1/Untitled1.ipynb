```python
import pandas as pd

# Load your dataset
df = pd.read_csv("C:/Users/S.B. Engineering/Downloads/food_choices.csv")

# Preview first rows
print(df.head())

# Quick info about columns & missing values
print(df.info())
```

```python
# Drop timestamp if present
df = df.drop(columns=["Timestamp"], errors="ignore")

# Fill missing with "Not specified" (for text) or 0 (for numbers)
df = df.fillna(0)
```

```python
print(df.info())
```

```python
# Summary of numeric columns
print(df.describe())

# Frequency counts for key categorical questions
print(df['Diet type'].value_counts())
print(df['Fast food frequency'].value_counts())
print(df['Check nutrition labels'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart: Fast food frequency
sns.countplot(x="Fast food frequency", data=df)
plt.title("Fast Food Consumption Frequency")
plt.show()

# Pie chart: Diet type
df['Diet type'].value_counts().plot.pie(autopct="%1.1f%%")
plt.title("Diet Type Distribution")
plt.ylabel("")
plt.show()

# Histogram: BMI distribution (if you calculated BMI)
sns.histplot(df['BMI'], bins=10, kde=True)
plt.title("BMI Distribution")
plt.show()
```

```python
# Correlation matrix for numeric values
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
