```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Setup complete')
```

```python
spotify_filepath='C:/Users/S.B. Engineering/Downloads/spotify.csv'
spotify_data=pd.read_csv(spotify_filepath,index_col="Date",parse_dates=True)
spotify_data.head()
```

```python
spotify_data.tail()
```

```python
sns.lineplot(data=spotify_data)
plt.show()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of each song 
sns.lineplot(data=spotify_data)
```

```python
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data=spotify_data['Despacito'], label="Despacito")

# Add label for horizontal axis
plt.xlabel("Date")
plt.show()
```

```python
sns.barplot(x=spotify_data.index,y=spotify_data['Shape of You'])
plt.show()
```

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Setup complete')

flight_path='C:/Users/S.B. Engineering/Downloads/flight_delays.csv'
flight_data=pd.read_csv(flight_path,index_col="Month")
flight_data.head()
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=flight_data.index, y=flight_data['NK'])
plt.show()
```

```python
sns.heatmap(data=flight_data, annot=True)
plt.show()
```

```python
sns.scatterplot(x=flight_data['AA'], y=flight_data['MQ'])
plt.show()
```

```python
sns.regplot(x=flight_data['AA'], y=flight_data['MQ'])
plt.show()
```
